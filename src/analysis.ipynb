```python
# Imports
from pattern.en import sentiment
import os
import numpy as np
```

```python
# First file test
with open("Z:/project-2025-sentiment/data/split/page_1.txt", "r") as file:
    text = file.read()
    for line in text.split("\n"):
        if line.endswith("-"):
            # Remove the -, then concatenate the next line
            text = text.replace(
                line,
                f"{line[:-1]} "
                + text.split("\n")[text.split("\n").index(line) + 1],
            )
    print(sentiment(text))
```

```python
sentiments = []
for file in os.listdir("Z:/project-2025-sentiment/data/split"):
    with open(f"Z:/project-2025-sentiment/data/split/{file}", "r") as file:
        text = file.read()
        for line in text.split("\n"):
            if line.endswith("-"):
                # Remove the -, then concatenate the next line
                text = text.replace(
                    line,
                    f"{line[:-1]} "
                    + text.split("\n")[text.split("\n").index(line) + 1],
                )
        sentiments.append(sentiment(text))

sentiments
```

```python
polarities = [sent[0] for sent in sentiments]
subjectivities = [sent[1] for sent in sentiments]

print(np.mean(polarities))
print(np.mean(subjectivities))
```
